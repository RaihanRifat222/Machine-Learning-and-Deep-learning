## In linnear regression, we find the best fit line to predict continuous numeric values. We measure the performance by calculating MSE. We use **Gradient Descent** to minimize the error. image.png

```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
house_data = pd.read_csv('/content/house_prices.csv')
house_data.head()
```

```python
size = house_data['sqft_living']
size
```

```python
price = house_data['price']
price
```

```python
x = np.array(size).reshape(-1,1)
y = np.array(price).reshape(-1,1)
x,y
```

Notice that the index values are gone now. This is because we have reshaped the dataframe to numpy arrays.

```python
model = LinearRegression()
model.fit(x,y)
```

```python
mse = mean_squared_error(x,y)
print(math.sqrt(mse))
```

```python
r_squared = model.score(x,y)
r_squared
```

```python
#This is the slope of the Line
print(model.coef_[0])
```

```python
#This is the intercept
model.intercept_[0]
```

```python
#We will see the line that we predicted
plt.scatter(x,y, color = 'green')
plt.plot(x, model.predict(x), color = 'black')
plt.title('Linear Regression')
plt.xlabel('Size')
plt.ylabel('Price')
plt.show()
```

```python
print(model.predict([[2000]]))
```
