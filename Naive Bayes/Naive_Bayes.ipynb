```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

data = pd.read_csv('/content/credit_data.csv')

features = data[['income', 'age', 'loan']]

target = data.default

trainX, testX, trainY, testY = train_test_split(features, target, test_size=0.3)
```

```python
features.corr()
```

```python

model = GaussianNB()
model.fit(trainX, trainY)
predictions = model.predict(testX)


```

```python
confusion_matrix(testY, predictions)
```

```python
accuracy_score(testY, predictions)
```

# Text Classification with dataset

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

categories = ['alt.atheism', 'soc.religion.christian', 'comp.graphics', 'sci.med']
training_data = fetch_20newsgroups(subset = 'train', categories = categories, shuffle = True, random_state= 3)

training_data
```

## We are printing the first 10 line of the first data.

```python
print('\n'.join(training_data.data[0].split('\n')[:10]))
```

## Now we will print thr target value of the first data

```python
print('Target is:', training_data.target_names[training_data.target[0]])
```

## Similarly, for the 2nd data

```python
print('\n'.join(training_data.data[1].split('\n')[:10]))
print('Target is:', training_data.target_names[training_data.target[1]])
```

## Now, we will train the text data and try to find out the target values of the texts.

### At first, we will tokenize the sentences.

#### For example,

I love Machine Learning.
I don't like Biology.

### After tokenizing. the first sentence is,
I - 1
love - 1
Machine - 1
Learning - 1
don't - 0
like - 0,
Biology - 0

## By using count_vector.vocabulary_ we can count the nummber of occurences of each word. In this case, the most used word is from.

```python
count_vector = CountVectorizer()
x_train_counts = count_vector.fit_transform(training_data.data)
print(count_vector.vocabulary_)
```

```python
x_train_counts
```

```python
tfid_Transformer  = TfidfTransformer()
x_train_tfidf = tfid_Transformer.fit_transform(x_train_counts)
x_train_tfidf
```

```python
model = MultinomialNB().fit(x_train_tfidf, training_data.target)

new = [' This has nothing to do with church of religion.', 'Software engineering is getting hotter']

x_new_counts = count_vector.transform(new)

x_new_tfidf = tfid_Transformer.transform(x_new_counts)

predicted = model.predict(x_new_tfidf)


for doc, category in zip(new, predicted):
  print('%r ------> %s' % (doc, training_data.target_names[category]))
```

# While doing text classification with Naive Bayes, why do we train the model with the transformed TFIDF version?

### When training a text classification model, using the transformed TF-IDF version is beneficial for several reasons:

## Feature Representation: TF-IDF provides a more informative representation of the text data. It captures the importance of each term (word) within a document relative to its frequency across the entire corpus.
## Dimensionality Reduction: By transforming the raw text into TF-IDF vectors, we reduce the dimensionality of the feature space. This helps mitigate the curse of dimensionality and improves model efficiency.
## Weighting Rare Terms: TF-IDF assigns higher weights to terms that are rare across documents. Rare terms often carry more discriminative power for classification tasks.
## Noise Reduction: Common words (e.g., the, and) have low TF-IDF scores due to their high document frequency. By downweighting these terms, we reduce noise in the feature space.
## Naive Bayes Assumption: Naive Bayes assumes that features are conditionally independent given the class label. While this assumption is rarely met in practice, using TF-IDF helps mitigate dependencies by focusing on term importance rather than raw counts.
